• Create a function that prompts the user for a task description. Assign a

```python
import json

# File to store tasks
TASK_DATA_FILE = 'tasks.json'

user_authentication = User_authentication()

class Task_management:
    # Function to list all tasks
    

    # Function to load task data from file
    def load_task_data(self):
        if os.path.exists(TASK_DATA_FILE):
            with open(TASK_DATA_FILE, 'r') as file:
                return json.load(file)   
        return {}

    # Function to save task data to file
    def save_task_data(self,task_data):
        with open(TASK_DATA_FILE, 'w') as file:
            json.dump(task_data, file)

    # Function to add a new task
    def add_task(self):
        task_data = self.load_task_data()
        task_description = input("Enter task description: ")
        task_id = str(len(task_data) + 1)
        task_status = "Pending"
        task_data[task_id] = {'description': task_description, 'status': task_status}
        self.save_task_data(task_data)
        print("Task added successfully!")
        
    # Function to view tasks for the logged-in user
    
    def view_tasks(self,user_authentication):
        user_data = user_authentication.load_user_data()
        username = input("Enter your username: ")
        if username not in user_data:
            print("Username not found. Please register first.")
            return
        password = input("Enter your password: ")
        hashed_password = user_authentication.hash_password(password)
        if user_data[username] != hashed_password:
            print("Invalid password. Please try again.")
            return
        
        task_data = self.load_task_data()
        print(f"Tasks for {username}:")
        for task_id, task_info in task_data.items():
            print(f"Task ID: {task_id}")
            print(f"Description: {task_info['description']}")
            print(f"Status: {task_info['status']}")
            print("")
            
    # Function to update the status of a task
    # Function to mark a task as completed
    def mark_task_completed(self):
        task_data = self.load_task_data()
        task_id = input("Enter the task ID to mark as completed: ")
        if task_id in task_data:
            task_data[task_id]['status'] = "Completed"
            self.save_task_data(task_data)
            print("Task marked as completed!")
        else:
            print("Task ID not found. Please try again.")
            
    # Function to delete a task
    
    def delete_task(self):
        task_data = self.load_task_data()
        task_id = input("Enter the task ID to delete: ")
        if task_id in task_data:
            del task_data[task_id]
            self.save_task_data(task_data)
            print("Task deleted successfully!")
        else:
            print("Task ID not found. Please try again.")

# Example usage
# add_task()
```

```python
# Function to delete a task
def delete_task(self):
    task_data = self.load_task_data()
    task_id = input("Enter the task ID to delete: ")
    if task_id in task_data:
        del task_data[task_id]
        self.save_task_data(task_data)
        print("Task deleted successfully!")
    else:
        print("Task ID not found. Please try again.")

# Example usage
# delete_task()
```

6. Create an Interactive Menu: 
• Build a menu that allows users to choose between: 
o Add a Task 
o View Tasks 
o Mark a Task as Completed 
o Delete a Task 
o Logout 
For each option, call the corresponding function, and loop back to the menu until 
the user logs out.

```python



user_auth = User_authentication()
task_mgmt = Task_management()

def interactive_menu():

    while True:
 
        
        
        print("\nTask Manager Menu:")
        print("1. Register user")
        print("2. Add a Task")
        print("3. View Tasks")
        print("4. Mark a Task as Completed")
        print("5. Delete a Task")
        print("6. Logout")
        
        choice = input("Enter your choice (1-6): ")
        print("############################################")
        
        if choice == '1':
            user_auth.register_user()
        elif choice == '2':
            task_mgmt.add_task()
        elif choice == '3':
            task_mgmt.view_tasks(user_auth)
        elif choice == '4':
           task_mgmt.mark_task_completed()
        elif choice == '5':
           task_mgmt. delete_task()
        elif choice == '6':
            print("Logging out...")
            break
        else:
            print("Invalid choice. Please try again.")

# Example usage

interactive_menu()
```
